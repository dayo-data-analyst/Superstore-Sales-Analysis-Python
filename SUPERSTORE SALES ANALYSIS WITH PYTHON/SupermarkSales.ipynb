**Univarante Analysis**
- Total Sales
- Total Quantity Sold
- Total Discount
- Total Profit

```python
total_sales = df['Sales'].sum()
total_quantity = df['Quantity'].sum()
total_discount = df['Discount'].sum()
total_profit = df['Profit'].sum()

print(f"Total Sales: {total_sales}")
print(f"Total Quantity: {total_quantity}")
print(f"Total Discount: {total_discount}")
print(f"Total Profit: {total_profit}")
```

```python
df.head()
```

```python
total_sales_by_region = df.groupby('Region')['Sales'].sum()
total_sales_by_region
plt.figure(figsize=(7,6))
total_sales_by_region.plot(kind='pie', title='Total Sales by Region', autopct='%1.1f%%')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
states_by_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
states_by_sales
states_by_sales.plot(kind='bar', title='Total Sales by State', figsize=(10,6), color='green')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```

```python
citysales = df.groupby('City')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
citysales.head(50)
```

```python
city_sales = df.groupby('City')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)[:10]

city_sales.plot(kind='bar', title='Total Sales by City', figsize=(10,6), color={'Sales':'orange', 'Profit':'green'}, stacked = True)
plt.xlabel('City')
plt.ylabel('Total Sales and Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
city_sales_by_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=True)[:10]

city_sales_by_profit.plot(kind='bar', title='Bottom Profit by City (Top 10)', figsize=(10,6), color='green')
plt.xlabel('City')
plt.ylabel('Bottom Profit')
plt.xticks(rotation=45)
plt.show()
```
